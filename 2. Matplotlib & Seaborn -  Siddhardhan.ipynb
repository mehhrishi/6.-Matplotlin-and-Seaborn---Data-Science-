Bar plot

```python
fig = plt.figure() #this will create empty plot
# and in that empty plot will do several things
# 'ax' is nothing but their axes
ax = fig.add_axes([0,0,1,1])
# this will enclose our plot in a rectangle
# 0,0 - is the origin
# 1,1 - height and width of the rectangle
languages=['English','French','Spanish','Latin','Germany']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('no. of people')
plt.title("Bar Plot")
plt.grid()
plt.show()
```

Pie chart 

```python
fig = plt.figure() #this will create empty plot
# and in that empty plot will do several things
# 'ax' is nothing but their axes
ax = fig.add_axes([0,0,1,1])
# this will enclose our plot in a rectangle
# 0,0 - is the origin
# 1,1 - height and width of the rectangle
languages=['English','French','Spanish','Latin','Germany']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

lets see, how we can build 3D plot

3D Scatter Plot

```python
# as this is 3d plot, will be using 3 axes - x,y,z 
fig=plt.figure()
ax=plt.axes(projection='3d')
z=20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,c=z,cmap='Blues')
plt.show()
```

Correlation: Basically there are two types of Correlation.

1.Postive Correlation

2.Negative Correlation

See, there are total 14 columns. and if i talk correlation between these two columns like crime rate and price. These both are negatively correlated, because, if the crime rate increases the rate of price decreases.

Negative Correlation: if one variable increases and another one variable decreases - then Negative Correlation

Positive Correlation: if one variable increases and another one also increases then it is said to be - Positive Correlation.

Postive Correlation examples: If the number of rooms increases then the price will also increases. In this case, no. of rooms and Price are Positively Correlated.

Negative Correlation: In Case of Crime and Price both these are Negatively Correlated.

### Heat Map

Constructing the Heat MAp

```python
# before this will find the correlation 
cor=df.corr()
cor
```

```python
# Constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(cor,cbar=True, square=True,fmt='.1f',annot=True,
            annot_kws={'size':8}, cmap="Blues")
```
