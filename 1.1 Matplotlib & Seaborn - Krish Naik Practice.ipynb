```python
#  compute the x and y coordinates for points on a sine curve 
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine wave Form")

# plot the points using matplotlib
plt.plot(x,y)
```

```python
# subplot
# compute the x and y coordinates for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
ysin=(np.sin(x))
ycos=(np.cos(x))

plt.subplot(2,1,1)

# make the first plot
plt.plot(x,ysin,'r--')
plt.plot("Sine")

# set the second subplot as active, and make the second plot
plt.subplot(2,1,2)
plt.plot(x,ycos,'g--')
plt.title("Cosine")

# show the figure
plt.show()
```

#### Join Plot
A join plot allows to study the relationship between 2 numeric variables(2 numerics datatype features). The central chart display correlation. it is usually a scatterplot, a hexbin plot, a 2D Histogram or a 2D density plot

Jopinplot: It actually helps you to do univariate analysis and when i say univariate again here we are taking 2 features, so we can also Bivariate analysis

But make sure that you just have 2 features while working with joinplot.

```python
#### Univariate Analysis
df.head()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#### Note: Above plot can be called as a Bivariatre Analysis because we have used 2 features

#### Now, if your features are more than 2, then will use Pair plot

### Pair plot
 - A "Pair Plot" is also known as Scatter Plot, in which one variabble in the same data row is matched with another variables's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all other variables.
 - In Pair plot, it will do Permutation and Combination of all the features and alway remember that when you are plotting this Pair plot, the features need to be of float or integer value

```python
sns.pairplot(df)
```

```python
# All the points which are plotted is based on the sex features.
# where blue is male
# orange is female
sns.pairplot(df,hue='sex')
```
